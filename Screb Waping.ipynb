```python
from requests import get # import our library
from bs4 import BeautifulSoup #import beautiful soup 
url = 'https://www.imdb.com/title/tt0137523/?ref_=ttls_li_i' #website we will be scraping
response = get(url) #pull information in the dataframe
html_soup = BeautifulSoup(response.text, 'html.parser') #parse text
print(response.text[:500]) #print first 500 characters in the text
```

```python
tables = html_soup.find_all(class_='article')
len(tables)
```

```python
tables=str(tables)
```

```python
import pandas as pd
dfs = pd.read_html(tables)
```

```python
billing = dfs[0]
billing
```

```python
billing = billing.drop([0,2], axis='columns')
```
